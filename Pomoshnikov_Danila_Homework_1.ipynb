# Central Bank of Russia Key Rate GAME

```python
import time 
from IPython.display import clear_output #очистка консоли
import random
```

```python
def CBRgame():
    game = True
    secondgame = False
    keyrate = [(15, 15.5), (15.5, 16), (16, 16.5), (16.5, 17), (17, 17.5)]

    while game == True:
        if secondgame == False:
            print('Привет, мой юный хомяк. Готов слить все свои деньги на бирже? Поехали!', flush = True)
            time.sleep(3)
            print('Сыграем в игру "Угадай, ключевую ставку ЦБ РФ в пятницу."', flush = True)
            time.sleep(3)
            clear_output(wait = True)
            print('Напишите "Да" или "Нет"', flush = True)
            answer = input()
            clear_output(wait = True)
            if answer == 'Нет':
                print('Игра окончена, ты сделал правильный выбор. Пора заняться своей жизнью.', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                break
            elif answer == 'Да':
                print('Глупец. Ты не знаешь, во что ввязался...', flush = True)
                time.sleep(2)
                clear_output(wait = True)
            else:
                print('Я не понял тебя.', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                continue

        mychoice = random.choice(keyrate)
        print('Отлично, я знаю, какую ставку установит ЦБ в пятницу. Попробой отгадать.', flush = True)
        clear_output(wait = True)
        while game == True:
            guess = float(input())
            if mychoice[0] <= guess < mychoice[1]:
                print('Поздравляю, ты отгадал! Самое время залететь на биржу с плечом и оставить там все свои сбережения.', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                print('Сыграть ещё раз?', flush = True)
                onemore = input()
                if onemore == 'Да':
                    print('Хорошо, если тебе не хватило, можем повторить.', flush = True)
                    time.sleep(2)
                    clear_output(wait = True)
                    secondgame = True
                    break
                elif onemore == 'Нет':
                    print('До свидания. Не повторяй больше данную ошибку...', flush = True)
                    time.sleep(2)
                    clear_output(wait = True)
                    game = False
                    break
                else:
                    print('Je ne te comprends pas. Еще раз.', flush = True)
                    time.sleep(2)
                    clear_output(wait = True)
                    continue
            else:
                print('Не, не угадал. Давай ещё раз.', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                continue
    print('Игра закончилась')     
```

## Запустить игру:

```python
CBRgame()
```
