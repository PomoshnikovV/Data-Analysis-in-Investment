```python
import time 
from IPython.display import clear_output #очистка консоли
import random

def pampgame():
    game = True
    secondgame = False
    pamp = ['Байков', 'Истомин', 'Шаклеина', 'Виноградов', 'Неклюдов', 'Худайкулова', 'Болгова']

    while game == True:
        if secondgame == False:
            print('Я искуственный интеллект, заместитель Игоря Александровича. \nСыграем в игру "Угадай сотрудника ПАМПА. Приз - 7-й мегатренд. Вы готовы???"', flush = True)
            time.sleep(2)
            clear_output(wait = True)
            print('Напишите "Да" или "Нет"', flush = True)
            answer = input()
            clear_output(wait = True)
            if answer == 'Нет':
                print('Игра окончена, вы сосланы в вышку', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                break
            elif answer == 'Да':
                print('Отлично, настоящий последователь мирового порядка', flush = True)
                time.sleep(2)
                clear_output(wait = True)
            else:
                print('То, что ты написал, даже хуже твоей курсовой. Игра окончена', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                break
        person = random.choice(pamp)
        print('Отлично, я выбрал памповца. Угадывай.', flush = True)
        clear_output(wait = True)
        while game == True:
            guess = input()
            if guess == person:
                print('Я поздравляю, вы победили', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                print('Сыграть ещё раз?', flush = True)
                onemore = input()
                if onemore == 'Да':
                    print('Гениально', flush = True)
                    time.sleep(2)
                    clear_output(wait = True)
                    secondgame = True
                    break
                elif onemore == 'Нет':
                    print('Ты теперь регионовед. хахахахахаха в рифму', flush = True)
                    time.sleep(2)
                    clear_output(wait = True)
                    game = False
                    break
                else:
                    print('Ты сейчас по-французски что-то сказал?', flush = True)
                    time.sleep(2)
                    clear_output(wait = True)
                    game = False
                    break
            else:
                print('Не, не угадал. Давай ещё раз.', flush = True)
                time.sleep(2)
                clear_output(wait = True)
                continue
    print('Игры закончились')     
    
```
