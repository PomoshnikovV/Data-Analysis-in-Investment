### Оператор if и логические операторы

Работает также, как в языке R, но немного отличается синтаксис: вложенная инструкция не в фигурных скобках, а после двоеточия.

X **and** Y

X **or** Y

**not** X

```python
a = int(input())
if a < -5:
    print('Слабо')
elif -5 <= a <= 5:
    print('Норм')
else:
    print('Сильно')
```

Для сокращения нотации можно использовать трехместную конструкцию else if

Как это бы выглядело с обычным **else if**:

if X:
    A = Y
else:
    A = Z

А как с **трехместной**:

A = Y if X else Z

### Циклы, операторы циклов

Циклы реализованы очень похоже, базово работаем с **for** и **while**. Как и с конструкцией условия, показателем вложенности является **отступ**, забыли про фигурные скобки.

Операторы позволяют комбинировать циклы с условиями, пропускать некоторые этапы итераций.

**continue** - пропускает оставшееся тело цикла и переходит к следуюзей итерации.

**break** - останавливает цикл.

!!!!!!!!!!!!!!!!!!!!!!!!!
**НОВЫЙ ДЛЯ ВАС ОПЕРАТОР**
!!!!!!!!!!!!!!!!!!!!!!!!!

**else** - проверяет, закончился ли цикл с помощью **break** или же закончился сам по себе

```python
i = 5
while i < 15:
    print(i)
    i = i + 2
```

```python
for i in 'статистические методы и регрессионный анализ':
    print(i * 2, end='')
```

```python
for i in 'Виктор Чадкин':
    if i == 'а' or i == 'о':
        print(i, end='')
        continue
    print(i * 4, end='')
```

```python
for i in 'Илья Орех':
    if i == 'а':
        break
else:
    print('Буквы a в строке нет')
```
