```python
import matplotlib.pyplot as plt

fig = plt.figure()   # Создание объекта Figure
print(fig.axes)   # Список текущих областей рисования пуст
print(type(fig))   # тип объекта Figure
plt.scatter(1.0, 1.0)   # scatter - метод для нанесения маркера в точке (1.0, 1.0)

# После нанесения графического элемента в виде маркера
# список текущих областей состоит из одной области
print(fig.axes)

plt.show()
```

```python
fig = plt.figure()
# Добавление на рисунок прямоугольной (по умолчанию) области рисования
ax = fig.add_axes([0, 0, 1, 1])
print(type(ax))
plt.scatter(1.0, 1.0)

fig = plt.figure()
# Добавление на рисунок круговой области рисования
ax = fig.add_axes([0, 0, 1, 1], polar=True)
plt.scatter(0.0, 0.5)


plt.show()
```

```python
#диаграмма
#требуется сочетание объект-количество
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7] 

#простой барплот
plt.bar(x, y)

#название
plt.title("Барплот")

#легенда
plt.legend(["бар"])
plt.show()
```

```python
#создадим данные
#оптимально использовать словарь
data = {'ежик':20, 'ёжик':15, 'еж':30, 
        'ёж':35}
ezhi = list(data.keys())
values = list(data.values())


#размер графика 
fig = plt.figure(figsize = (10, 5))

#барплот
plt.bar(ezhi, values, color ='maroon', 
        width = 0.4)

plt.xlabel("ежи")
plt.ylabel("количество")
plt.title("еееееежи")
plt.show()
```

```python
#создание данных
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]

#размер графика 
fig = plt.figure(figsize = (10, 5))

#точечный график
plt.scatter(x, y)

#легенда
plt.legend("A")

#заголовок
plt.title("точки")
plt.show()
```

```python
#стек-график
#показывает структуру объекта
days = [1, 2, 3, 4, 5]

#часы на питон
studying = [7, 8, 6, 11, 7]

#часы на хоечку
hoi = [8, 5, 7, 8, 13]

#рисуем стекплот
plt.stackplot(days, studying, hoi,
            colors =['r', 'c'])

#подписываем ось x
plt.xlabel('Дни')

#теперь у
plt.ylabel('№')

# Title of Graph
plt.title('Соотношение игры в хоечку и учебы')

#отображаем график
plt.show()
```

```python
#усы
import numpy as np

#генерируем данные
np.random.seed(10)
data = np.random.normal(100, 20, 200)

fig = plt.figure(figsize =(10, 7))

#рисуем
plt.boxplot(data)

#показываем
plt.show()
```

```python
#создаем данные
x = [1, 2, 3, 4] 

#данный параметр задаст расстояние между секциями
e  =(0.1, 0.1, 0.1, 0.1)

#рисуем
plt.pie(x, explode = e)

#подписываем и показываем
plt.title("пирог")
plt.show()
```
